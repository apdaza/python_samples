```python
def crear_carpeta_facultad(facultad):
    srcgen_folder = facultad 
    if not exists(srcgen_folder):
        mkdir(srcgen_folder)
```

```python
def crear_carpeta(facultad, id):
    srcgen_folder = facultad + "/" + id
    if not exists(srcgen_folder):
        mkdir(srcgen_folder)
```

```python
def convert_url_to_pdf(url, pdf_path):
    try:
        pdfkit.from_url(url, pdf_path, options={"enable-local-file-access": ""})
        print(f"PDF generated and saved at {pdf_path}")
    except Exception as e:
        print(f"PDF generation failed: {e}")
```

```python
def crear_html(estudiante):
    with open(join(estudiante['FACULTAD']+"/"+str(estudiante['IDENTIFICACION']), "%s_ha.html" % estudiante['IDENTIFICACION']), 'w') as f: f.write(template.render(entity=estudiante))
    
```

```python
def crear_pdf(estudiante):
    nombre = estudiante['FACULTAD']+"/"+str(estudiante['IDENTIFICACION'])+"/"+str(estudiante['IDENTIFICACION'])+"_ha"
    if not exists(nombre+".pdf"):
        convert_url_to_pdf(nombre+".html", nombre+".pdf")
    else:
        print("ya existe el archivo: " + nombre+".pdf")
```

```python
def borrar_html(estudiante):
    nombre = estudiante['FACULTAD']+"/"+str(estudiante['IDENTIFICACION'])+"/"+str(estudiante['IDENTIFICACION'])+"_ha"
    remove(nombre+".html")
```

```python
for i in range(len(df)):
    crear_carpeta_facultad(df.iloc[i]['FACULTAD'])
    crear_carpeta(df.iloc[i]['FACULTAD'],str(df.iloc[i]['IDENTIFICACION']))
```
